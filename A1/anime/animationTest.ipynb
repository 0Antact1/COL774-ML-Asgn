```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as anime
```

```python
fig, ax = plt.subplots()

x1 = np.random.normal(0,4,20)
x2 = np.random.normal(5,2,20)

x1_prog = []
x2_prog = []

ax.set_xlim(-10,10)
ax.set_ylim(2,12)
writer = anime.PillowWriter(fps=10)

with writer.saving(fig, "scatterAnime.gif", 100):
    for i in range(20):
        x1_prog.append(x1[i])
        x2_prog.append(x2[i])

        ax.scatter(x1_prog,x2_prog)
        writer.grab_frame()
```
